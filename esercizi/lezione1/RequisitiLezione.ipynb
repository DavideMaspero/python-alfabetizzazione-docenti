# LEZIONE 1 - Cosa è necessario conoscere

1. Cos'è una variabile e come le si assegna un valore (tramite un'istruzione di assegnamento)
        variabile1 = 100
        variabile2 = 10.2
        variabile3 = 'Hello world!'
        variabile4 = True
1. L'operatore di divisione `/` (tra due valori numerici)
        divisione = variabile1 / variabile2
1. L'operatore di divisione intera `//` (tra due valori numerici)
        divisione = variabile1 // variabile2
1. L'operatore di resto della divisione intera `%` (tra due valori numerici)
        resto_divisione_intera = variabile1 % variabile2
1. L'operatore di somma `+` (tra due valori numerici)
        somma = variabile1 + variabile2
1. L'operatore di prodotto `*` (tra due valori numerici)
        prodotto = variabile1 * variabile2
1. L'operatore di elevamento a potenza `**` (tra due valori numerici)
        quadrato = variabile1**2
        radice_quadrata = variabile1**0.5
1. L'operatore di negazione aritmetica `-`
        negazione_aritmetica = -variabile1
1. Gli operatori di confronto `>` e `<` (usati in un'espressione che restituisce `True` o `False`)
          variabile2 < variabile1
          variabile2 > variabile1
1. Cos'è una stringa di caratteri e come si assegna una stringa a una variabile
        stringa = 'Hello world!'
1. La funzione predefinita `len()` per determinare la dimensione di una stringa
        dimensione_stringa = len(stringa)
1. L'indicizzazione positiva dei caratteri di una stringa: 0 è l'indice del primo carattere, 1 quello del secondo, etc., `dimensione_stringa-1` è l'indice dell'ultimo carattere)
1. L'indicizzazione negativa dei caratteri di una stringa: -1 è l'indice dell'ultimo carattere, -2 quello del penultimo, etc., `-dimensione_stringa` è l'indice del primo carattere
1. La sintassi di accesso al singolo carattere di una stringa tramite l'indice della sua posizione
        primo_carattere = stringa[0]
        primo_carattere = stringa[-dimensione_stringa]
1. La sintassi di *slicing* per accedere a una sottostringa
        sottostringa_da_terzo_a_ottavo_carattere = stringa[2:8]
1. L'operatore `in` per testare la presenza di un carattere all'interno di una stringa (usato in un'espressione che restituisce `True` of `False`)
        'w' in stringa
1. Cos'è una lista e come si assegna una lista a una variabile
        lista = [False, 100, 345.789, 'Hello world!']
1. La funzione predefinita `len()` per determinare la dimensione di una lista
        dimensione_lista = len(lista)
1. L'indicizzazione positiva degli elementi di una lista: 0 è l'indice del primo elemento, 1 quello del secondo, etc., `dimensione_lista-1` è l'indice dell'ultimo elemento)
1. L'indicizzazione negativa degli elementi di una lista: -1 è l'indice dell'ultimo elemento, -2 quello del penultimo, etc., `-dimensione_lista` è l'indice del primo elemento
1. La sintassi di accesso al singolo elemento di una lista tramite l'indice della sua posizione
        terzo_elemento = lista[2]
        terzo_elemento = lista[-dimensione_lista+2]
1. La sintassi di aggiornamento del singolo elemento di una lista tramite l'indice della sua posizione
        lista[2] = 100
1. La sintassi di *slicing* per accedere a una sottolista
        sottolista_da_secondo_a_terzo_elemento = lista[1:3]
1. La sintassi di aggiornamento di una sottolista tramite *slicing*
        lista[1:3] = [10,10,10,10]
1. L'operatore `in` per testare la presenza di una chiave all'interno di un dizionario (usato in un'espressione che restituisce `True` of `False`)
        456 in lista
1. Cos'è un dizionario e come si assegna un dizionario a una variabile
        dizionario = {'mele' : 4, 'pere' : 6, 'fragole' : 15}
1. La funzione predefinita `len()` per determinare il numero di coppie chiave-valore di un dizionario
        dimensione_dizionario = len(dizionario)
1. La sintassi di accesso al singolo valore tramite la sua chiave
        numero_mele = dizionario['mele']
1. La sintassi di aggiornamento del singolo valore tramite la sua chiave
        dizionario['mele'] = 6
1. L'operatore `in` per testare la presenza di una chiave all'interno di un dizionario (usato in un'espressione che restituisce `True` of `False`)
        'ciliegie' in dizionario       
1. Il costrutto di selezione `if`
        if condizione_da_verificare_vera:
            faccio_qualcosa
1. Il costrutto di selezione `if-else` (che può essere anche annidato)
        if condizione_da_verificare_vera:
            faccio_qualcosa
        else:
            faccio_altro   
1. Il costrutto di selezione `elif`
        if condizione1_da_verificare_vera:
            eseguo_task1
        elif condizione2_da_verificare_vera:
            eseguo_task2
        elif condizione3_da_verificare_vera:
            eseguo_task3
        ...
        elif condizioneN_da_verificare_vera:
            eseguo_taskN
        else:
            eseguo_task_finale
1. La sintassi scansione degli elementi di una lista tramite `for`
        for elemento in lista:
            faccio_qualcosa
